```python
import nltk
import math
import os
import random
# Import necessary Python libraries: `nltk` for natural language processing, `math` for mathematical operations, `os` for interacting with the operating system, and `random` for generating random numbers.

from nltk.lm.preprocessing import padded_everygram_pipeline
from nltk.lm import Laplace
from nltk.util import ngrams, pad_sequence
from nltk.corpus import brown
# Import specific functions and classes from the NLTK library: `padded_everygram_pipeline` for preparing data for language modeling, `Laplace` for smoothing in language models, `ngrams` and `pad_sequence` for creating n-grams, and `brown` for accessing the Brown corpus.

import google.generativeai as GeminiAI
from google.generativeai.types import HarmCategory, HarmBlockThreshold
# Import the `google.generativeai` library as `GeminiAI` for working with Google's generative AI tools, and import specific types `HarmCategory` and `HarmBlockThreshold` for handling harm-related configurations.

from google.colab import userdata
# Import `userdata` from `google.colab`, which is typically used for managing user data in Google Colab environments.

print("Downloading NLTK data (Brown corpus and Punkt tokenizer)...")
nltk.download('brown')
nltk.download('punkt')
print("Downloads complete.")

random.seed(42)
# Set the seed for the random number generator to 42 to ensure reproducibility of random operations.
```

```python
print("\nPart 1: Building Trigram (n=3) Language Model")

# Define the order of the n-gram model as 3 (trigram).
N_ORDER = 3

print("Loading and preprocessing Brown corpus (lowercasing)...")

# Convert all words in the Brown corpus to lowercase and store them in a list of sentences.
Corpus = [[word.lower() for word in sent] for sent in brown.sents()]

# Prepare the training data and vocabulary for the trigram model using padding.
TrainData, vocab = padded_everygram_pipeline(N_ORDER, Corpus)

# Initialize a Laplace (add-one smoothing) n-gram language model of order 3.
lm = Laplace(N_ORDER)

# Fit the language model to the training data and vocabulary.
lm.fit(TrainData, vocab)

# Print a message indicating the successful creation of the trigram Laplace model.
print(f"Successfully built a {N_ORDER}-gram (Trigram) Laplace model.")

# Print the size of the vocabulary used in the model.
print(f"Vocabulary size: {len(lm.vocab)}")
```

```python
print("\nPart 2: Simple Prediction (starting with 'i am')")

Words = ['i', 'am']
# Initialize a list called 'Words' with the seed words 'i' and 'am'.

Predicted = lm.generate(1, text_seed=Words, random_seed=42)[-1]
# Use a language model 'lm' to predict the next word based on the seed 'Words'.
# The prediction is deterministic due to the fixed 'random_seed'.

print(f"Seed context: '{' '.join(Words)}'")

print(f"Model prediction for the next word: '{Predicted}'")

print(f"Complete phrase: 'i am {Predicted}'")
```

```python
print("\nPart 3: Generating 10 Sentences (Start: 'you are')")

Start = ['you', 'are']
# Define the starting words for each sentence as a list of tokens.

GeneratedSentenceTokens = []
# Initialize an empty list to store the tokens of each generated sentence.

GeneratedSentencesStrings = []
# Initialize an empty list to store the string form of each generated sentence.

print("Generating 10 unique sentences...")

for i in range(10):
    # Loop 10 times to generate 10 sentences.

    SentenceToken = lm.generate(20, text_seed=Start, random_seed=i)
    # Generate a sequence of 20 tokens using a language model (lm), starting with the 'Start' tokens and using a different random seed for each iteration.

    FullToken = Start + SentenceToken
    # Combine the starting tokens with the generated tokens to form the full sentence in token form.

    GeneratedSentenceTokens.append(FullToken)
    # Append the full token list of the sentence to the list of generated sentence tokens.

    SentenceString = " ".join(FullToken)
    # Convert the list of tokens into a single string with spaces between tokens.

    GeneratedSentencesStrings.append(SentenceString)
    # Append the sentence string to the list of generated sentence strings.

    FinalSentenceString = SentenceString.capitalize()
    # Capitalize the first letter of the sentence string.

    if not FinalSentenceString.endswith('.'):
        FinalSentenceString += "."
    # Ensure the sentence ends with a period.

    print(f"{i+1}. {FinalSentenceString}")
    # Print the sentence number and the final formatted sentence string.
```

```python
print("\nPart 4: Calculating Sentence Probabilities")

SentenceMetrics = []
# Initialize a list to store metrics for each sentence

for i, tokens in enumerate(GeneratedSentenceTokens):
  # Iterate over each tokenized sentence

    PadLeftSymbols = ['<s>'] * (N_ORDER - 1)
    # Create padding symbols for the start of the sentence
    ScoreTokens = PadLeftSymbols + tokens + ['</s>']
    # Add start and end symbols to the sentence tokens

    NumTokens = len(tokens) + 1
    # Calculate the number of tokens including the end symbol

    SentenceNgrams = list(ngrams(ScoreTokens, N_ORDER))
    # Generate n-grams from the padded sentence

    TotalLogProbability = 0.0
    # Initialize the total log probability for the sentence
    for gram in SentenceNgrams:
      # Iterate over each n-gram
        Context = gram[:-1]
        # Extract the context (all but the last word of the n-gram)
        Word = gram[-1]
        # Extract the target word (last word of the n-gram)
        TotalLogProbability += lm.logscore(Word, Context)
        # Add the log probability of the word given the context

    SentenceMetrics.append((TotalLogProbability, NumTokens))
    # Store the log probability and number of tokens

    Probability = math.exp(TotalLogProbability)
    # Convert log probability to actual probability

    # Print the results for the current sentence
    print(f"\nSentence {i+1}: '{GeneratedSentencesStrings[i]}'")
    print(f"  Log-Probability: {TotalLogProbability:.4f}")
    print(f"  Probability: {Probability:.4e}")

# Explanation:
  # The code calculates the probability of each generated sentence using an n-gram language model.
  # It first pads each sentence with start (`<s>`) and end (`</s>`) symbols to handle sentence boundaries.
  # It then generates n-grams from the padded sentence and calculates the total log probability by summing the log probabilities of each word given its context.
  # The log probability is converted to actual probability using the exponential function.
  # Finally, it prints the log probability and actual probability for each sentence and stores these metrics for further analysis.
```

```python
print("\nPart 5: Calculating Model Perplexity")

# Initialize variables to accumulate total log-probability and token count
TotalCorpusLogProbability = 0.0
TotalCorpusTokens = 0

# Iterate over each sentence's log-probability and token count in SentenceMetrics
for log_p, n_tok in SentenceMetrics:
    # Accumulate the log-probability of the sentence to the total
    TotalCorpusLogProbability += log_p
    # Accumulate the number of tokens in the sentence to the total
    TotalCorpusTokens += n_tok

print(f"Total Log-Probability of all 10 sentences: {TotalCorpusLogProbability:.4f}")
print(f"Total number of tokens (N) including </s> markers: {TotalCorpusTokens}")

# Calculate the average cross-entropy by dividing the negative total log-probability by the total number of tokens
CrossEntropy = - (TotalCorpusLogProbability / TotalCorpusTokens)
print(f"Average Cross-Entropy (H): {CrossEntropy:.4f}")

# Calculate perplexity by exponentiating the cross-entropy
Perplexity = math.exp(CrossEntropy)

print(f"\nOverall Perplexity on the 10 generated sentences: {Perplexity:.4f}")

# Explanation:
  # The code calculates the perplexity of a language model based on the log-probabilities of sentences.
  # It accumulates the total log-probability and token count from a list of sentence metrics.
  # Cross-entropy is computed as the negative average log-probability per token.
  # Perplexity, a measure of how well the model predicts the test data, is derived by exponentiating the cross-entropy.
  # The results are printed, showing the total log-probability, token count, cross-entropy, and perplexity.
```

```python
print("\nPart 6: Generating Story with Google Gemini (and Validation)")

try:

    # Initialize an empty list to store cleaned sentences.
    CleanedSentences = []

    # Iterate over each sentence in the list of generated sentence strings.
    for s in GeneratedSentencesStrings:

        # Remove the "</s>" tag and strip leading/trailing whitespace.
        cleaned = s.replace("</s>", "").strip()

        # If the sentence ends with a period, remove it and strip whitespace again.
        if cleaned.endswith('.'):
            cleaned = cleaned[:-1].strip()

        # Add the cleaned sentence to the list.
        CleanedSentences.append(cleaned)

    # Retrieve the API key from user data.
    API_KEY = userdata.get('nle')

    # If the API key is not found, raise an error.
    if not API_KEY:
        raise ValueError("API Key 'nle' not found in Colab secrets.")

    # Configure the GeminiAI with the retrieved API key.
    GeminiAI.configure(api_key=API_KEY)

    print("Building the prompt for the Gemini API...")

    # Join the cleaned sentences into a single string with newline separation.
    PromptSentences = "\n".join(CleanedSentences)

    # Create a prompt for the Gemini API to generate a story.
    GeminiPrompt = f"""
    Write a short story (less than 500 words) that must contain
    all of the following 10 sentences exactly as written.
    Do not add any punctuation (like periods) to the end of them if they don't have it:

    {PromptSentences}

    Ensure the story flows well and is cohesive.
    The story MUST NOT contain any DEROGATORY, TOXICITY, VIOLENCE, HATE_SPEECH, SEXUAL, or HARASSMENT content.
    The story should be safe for all audiences.
    """

    print("PROMPT TO BE SENT TO GEMINI")
    print(GeminiPrompt)

    # Define safety settings to block high-risk harmful content.
    SafetySettings = {
        HarmCategory.HARM_CATEGORY_DANGEROUS_CONTENT: HarmBlockThreshold.BLOCK_ONLY_HIGH,
        HarmCategory.HARM_CATEGORY_HATE_SPEECH: HarmBlockThreshold.BLOCK_ONLY_HIGH,
        HarmCategory.HARM_CATEGORY_HARASSMENT: HarmBlockThreshold.BLOCK_ONLY_HIGH,
        HarmCategory.HARM_CATEGORY_SEXUALLY_EXPLICIT: HarmBlockThreshold.BLOCK_ONLY_HIGH,
    }

    # Initialize the GeminiAI generative model with specified safety settings.
    Model = GeminiAI.GenerativeModel('gemini-2.5-pro', safety_settings=SafetySettings)

    print("Sending prompt to Gemini 2.5 Pro API... This may take a moment.")

    # Generate content using the Gemini model with the constructed prompt.
    Response = Model.generate_content(GeminiPrompt)

    print("\nGenerated Story (from Gemini)")
    GeneratedStory = Response.text
    print(GeneratedStory)

    # Begin analysis and testing of the generated story.
    print("Story Analysis and Tests")

    # Calculate the word count of the generated story.
    WordCount = len(GeneratedStory.split())
    print(f"Total Word Count: {WordCount}")

    # Check if the story is under 500 words and print the result.
    if WordCount <= 500:
        print("TEST PASSED: Story is under 500 words.")
    else:
        print(f"TEST FAILED: Story is {WordCount} words (over 500).")

    # Initialize a list to track missing sentences.
    MissingSentences = []

    # Check if each cleaned sentence is present in the generated story.
    for sentence in CleanedSentences:
        if sentence not in GeneratedStory:
            MissingSentences.append(sentence)

    if not MissingSentences:
        print(f"TEST PASSED: All 10 required sentences were found.")
    else:
        print(f"TEST FAILED: {len(MissingSentences)} sentences are missing:")
        for s in MissingSentences:
            print(f"    - {s}")

    # Define a list of negative keywords to check for in the story.
    NEGATIVE_KEYWORDS = [
        'derogatory', 'toxicity', 'violence', 'hate_speech',
        'sexual', 'harassment'
    ]

    # Initialize a list to track found negative keywords.
    FoundKeywords = []

    # Convert the generated story to lowercase for keyword checking.
    StoryLower = GeneratedStory.lower()

    # Check for the presence of each negative keyword in the story.
    for keyword in NEGATIVE_KEYWORDS:
        if keyword in StoryLower:
            FoundKeywords.append(keyword)

    if not FoundKeywords:
        print(f"TEST PASSED: No forbidden keywords found.")
    else:
        print(f"TEST FAILED: Found {len(FoundKeywords)} forbidden keyword(s):")
        for k in FoundKeywords:
            print(f"    - {k}")

    print("TEST PASSED: Story was generated by the API.")
    print("   (This implies it passed the API-level SafetySettings filter.)")

# Handle any exceptions that occur during the process.
except Exception as e:
    print(f"\nAPI Call Failed")
    print(f"Error: {e}")
    print("\nPlease ensure you have a valid API key named 'nle' in your Colab secrets.")
    print("Also, check that you have access to the 'gemini-2.5-pro' model.")

    # Set default values indicating no story was generated.
    GeneratedStory = "API call failed. No story was generated."
    WordCount = 0
```

```python
print("\nSaving outputs to disk")

os.makedirs("assignment_outputs", exist_ok=True)
# Creates a directory named "assignment_outputs" if it doesn't already exist.

with open("assignment_outputs/sentences_and_metrics.txt", "w", encoding="utf-8") as f:
    # Opens a file named "sentences_and_metrics.txt" in write mode with UTF-8 encoding within the "assignment_outputs" directory.

    f.write(f"Overall Perplexity: {Perplexity:.4f}\n")
    # Writes the overall perplexity value to the file, formatted to four decimal places.

    f.write(f"Overall Cross-Entropy: {CrossEntropy:.4f}\n")
    # Writes the overall cross-entropy value to the file, formatted to four decimal places.

    f.write(f"Total Log-Probability: {TotalCorpusLogProbability:.4f}\n")
    # Writes the total log-probability of the corpus to the file, formatted to four decimal places.

    f.write(f"Total Tokens (N): {TotalCorpusTokens}\n\n")
    # Writes the total number of tokens in the corpus to the file, followed by a newline for separation.

    f.write("INDIVIDUAL SENTENCE METRICS\n\n")
    # Writes a header for the individual sentence metrics section.

    for i in range(len(GeneratedSentencesStrings)):
        # Iterates over each generated sentence.

        s = GeneratedSentencesStrings[i]
        # Retrieves the current sentence string.

        log_p, n_tok = SentenceMetrics[i]
        # Retrieves the log-probability and token count for the current sentence.

        Probablity = math.exp(log_p)
        # Calculates the probability of the sentence by exponentiating its log-probability.

        f.write(f"Sentence {i+1}: {s}\n")
        # Writes the sentence number and the sentence itself to the file.

        f.write(f"  Probability: {Probablity:.5e}\n")
        # Writes the probability of the sentence in scientific notation.

        f.write(f"  LogProb: {log_p:.5f}\n")
        # Writes the log-probability of the sentence, formatted to five decimal places.

        f.write(f"  Tokens (incl </s>): {n_tok}\n\n")
        # Writes the number of tokens in the sentence, including the end-of-sentence token, followed by a newline.

print("Saved sentences_and_metrics.txt to assignment_outputs/")
# Prints a message indicating that the "sentences_and_metrics.txt" file has been saved.

with open("assignment_outputs/generated_story.txt", "w", encoding="utf-8") as f:
    # Opens a file named "generated_story.txt" in write mode with UTF-8 encoding within the "assignment_outputs" directory.

    f.write(GeneratedStory)
    # Writes the generated story to the file.

print("Saved generated_story.txt to assignment_outputs/")

print("\nAssignment 1 Code Execution Complete")
```
